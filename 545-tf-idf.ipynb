```python
import numpy as np
import matplotlib.pyplot as plt
import csv
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
text_data = []
text_label = []

with open('/kaggle/input/feedback-prize-2021/train.csv', encoding='utf-8') as file_obj:
    reader = csv.reader(file_obj)
    for row in reader:
        text_data.append(row[4])
        text_label.append(row[5])

text_data.pop(0)
text_label.pop(0)
C_value = 1.9
X_train, X_test, y_train, y_test = train_test_split(text_data, text_label, test_size=0.3, random_state=42)
text_clf_svm = Pipeline([('vect', CountVectorizer()), ('tfidf', TfidfTransformer()),
                         ('clf-svm', LinearSVC(loss='hinge', penalty='l2', max_iter=15000, random_state=42, C=C_value))])
text_clf_svm = text_clf_svm.fit(X_train, y_train)
result = text_clf_svm.predict(X_test)
```
